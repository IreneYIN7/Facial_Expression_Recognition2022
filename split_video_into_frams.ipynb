```python
import cv2


def split_into_fram(video_path, dir_path):
  
  vidcap = cv2.VideoCapture(video_path)
  success,image = vidcap.read()
  currentFrame = 0

  vidcap.set(cv2.CAP_PROP_POS_MSEC,4000) # skipping directly to the 5th second in the video.
  num = 0
  while success:
    name = 'Target3_emotion' + str(currentFrame) + '.jpg'
    cv2.imwrite(dir_path + name, image)     # save frame as JPEG file
    success,image = vidcap.read()

    print('Read a new frame: ', success)
    if cv2.waitKey(10) == 27:                     # exit if Escape is hit
      break
    currentFrame += 1
```

```python
labels = ['fear', "happy", 'neutral', 'sad']
for emotion in labels:
  target_path = "/content/drive/MyDrive/2021_Zhebin_CSThesis/target1_exp4/" + emotion+".avi"
  dir_path = "/content/drive/MyDrive/2021_Zhebin_CSThesis/Experiment_data/target1_exp3/" + emotion+"/"
  split_into_fram(target_path, dir_path)
```
